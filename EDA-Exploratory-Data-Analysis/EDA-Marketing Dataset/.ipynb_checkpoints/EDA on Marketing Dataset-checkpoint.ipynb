```python
#import the useful libraries.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Read the data set of "Marketing Analysis" in data.
data= pd.read_csv("marketing_data.csv")

# Printing the data
print(data.info())
```

```python
data.head() #5 
```

```python
#Dropping customerid as it has no importance in the analysis

df = data.copy()
df = df.drop('customerid',axis=1)
```

```python
df.head()      #response is the target variable
```

```python
#splitting jobedu column as it contains job and educational details

#Extract job  & Education in newly from "jobedu" column.
df['job'] = df['jobedu'].apply(lambda x: x.split(",")[0])
df['education'] = df['jobedu'].apply(lambda x: x.split(",")[1])

# Drop the "jobedu" column from the dataframe.
df.drop('jobedu',axis = 1, inplace = True)

# Printing the Dataset
df.head()
```

```python
#checking the missing values
#df.isnull().sum()  
(df.isnull().sum()/45211)*100  #df.shape[0]
```

```python
#checking isna
df.isna().sum()
```

## Never ever impute/change target variable ie response in above data

```python
#Replacing missing values of age with median
age_median = df.age.median()  #sns.displot 
print(age_median)
df.age.fillna(age_median,inplace=True)
```

```python
df.isnull().sum()
```

```python
# Dropping the records with month missing in data dataframe.
df = df[~df.month.isnull()].copy()
df = df[~df.response.isnull()].copy()
#df = df[~df.month.isnull()]
```

```python
#Bivariate analysis
#plot the scatter plot of balance and salary variable in data
plt.scatter(df.salary,df.balance)

#plot the scatter plot of balance and age variable in data
df.plot.scatter(x="age",y="balance")
plt.show()
```

```python
#corr [-1,1]
# Creating a matrix using age, salry, balance as rows and columns
print(df[['age','salary','balance']].corr())  #[-1,1]

#plot the correlation matrix of salary, balance and age in data dataframe.
sns.heatmap(df[['age','salary','balance']].corr(), annot=True, cmap = 'Reds')
plt.show() # if the correlation value > 0.7
```

```python
#plot the correlation matrix of salary, balance and age in data dataframe.
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap = 'Reds')
plt.show()
```

```python
#categorical analysis
##groupby the response to find the mean of the salary with response no & yes separately.
print(df.groupby('response')['salary'].mean())
```

```python
#median
print(df.groupby('response')['salary'].median())
```

```python
#By both mean and median salary for Yes and No are same
#lets check box plot
sns.boxplot(data.response, data.salary)
plt.show()

#The IQR for customers who gave a positive response is on the higher salary side.
#displot (bins=10)
```

```python
#categorical-categorical analysis
#create response_rate of numerical data type where response "yes"= 1, "no"= 0
df['response_rate'] = np.where(df.response=='yes',1,0)
df.response_rate.value_counts()
```

```python
#plot the bar graph of marital status with average value of response_rate
df.groupby('marital')['response_rate'].mean().plot.bar()
plt.show()
```

```python
#multi-variate analysis
result = pd.pivot_table(data=df, index='education', columns='marital',values='response_rate')
print(result)

#create heat map of education vs marital vs response_rate
sns.heatmap(result, annot=True, cmap = 'RdYlGn', center=0.117)
plt.show()
```

```python
#multi-variate analysis
result = pd.pivot_table(data=df, index='job', columns='marital',values='response_rate')
print(result)

#create heat map of education vs marital vs response_rate
sns.heatmap(result, annot=True, cmap = 'RdYlGn', center=0.117)
plt.show()
```

```python
#multi-variate analysis
result = pd.pivot_table(data=df, index='education', columns='poutcome',values='response_rate')
print(result)

#create heat map of education vs marital vs response_rate
sns.heatmap(result, annot=True, cmap = 'RdYlGn', center=0.117)
plt.show()
```
